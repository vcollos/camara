```python
# Converta os dados em DATA para um formato específico
# Primeiro, vamos verificar o tipo atual dos dados na coluna DATA
print("Tipo dos dados em DATA antes da conversão:", type(df_export_csv['DATA'].iloc[0]))

# Forçar o formato de data correto
# Independentemente do formato atual, vamos converter para string no formato brasileiro
if isinstance(df_export_csv['DATA'].iloc[0], pd.Timestamp) or isinstance(df_export_csv['DATA'].iloc[0], datetime):
    # Se já for um objeto datetime ou Timestamp
    df_export_csv.loc[:, 'DATA'] = df_export_csv['DATA'].dt.strftime('%d/%m/%Y')
else:
    # Se for uma string ou outro formato
    df_export_csv.loc[:, 'DATA'] = pd.to_datetime(df_export_csv['DATA']).dt.strftime('%d/%m/%Y')

# Verifica se a conversão foi bem-sucedida
print("Primeiras linhas de DATA após conversão:", df_export_csv['DATA'].head())

import csv

# Formato brasileiro (vírgula como separador decimal)
def format_valor(val):
    # Garante que o valor está em formato float, mesmo se vier como string com vírgula
    if isinstance(val, str):
        val = val.replace(',', '.')
    return f'{float(val):.2f}'.replace('.', ',')

df_export_csv.loc[:, 'valor'] = df_export_csv['valor'].apply(format_valor)

# Verificação manual do DataFrame antes da exportação
print("\nVERIFICANDO DATAFRAME ANTES DA EXPORTAÇÃO:")
for col in df_export_csv.columns:
    print(f"Coluna {col} - tipo: {type(df_export_csv[col].iloc[0])}")
    print(f"Exemplo: {df_export_csv[col].iloc[0]}")

# Exporta sem aspas, escrevendo manualmente o arquivo
with open('exported_data.csv', 'w', encoding='utf-8', newline='') as f:
    # Escreve o cabeçalho
    f.write('Debito;Credito;Historico;DATA;valor;complemento\n')
    
    # Escreve cada linha sem aspas, com verificação explícita para cada coluna
    for _, row in df_export_csv.iterrows():
        # Força DATA a estar no formato dd/mm/yyyy se ainda não estiver
        data_formatada = row['DATA']
        if '00:00:00' in str(data_formatada):
            # Ainda está no formato errado, forçamos a mudança
            data_formatada = datetime.strptime(
                str(data_formatada).split(' ')[0], '%Y-%m-%d'
            ).strftime('%d/%m/%Y')
        
        linha = f"{row['Debito']};{row['Credito']};{row['Historico']};{data_formatada};{row['valor']};{row['complemento']}\n"
        f.write(linha)

print("Exportação concluída!")
```
